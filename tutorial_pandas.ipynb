# Tutorial de Pandas: Como crear y filtrar un DataFrame

```python
import pandas as pd
```

```python
# Manipular datos en pandas
datosDePaises = {"Mexico":[0, 0, 4], "China":[39, 30, 30]}
df_medallero2021 = pd.DataFrame.from_dict(datosDePaises)

# Agregar diccionario, llaves como filas
df_medalleroPaises = pd.DataFrame.from_dict(datosDePaises, orient="index", columns=["oro", "plata", "bronce"])
df_medalleroPaises
```

### DataFrame
Un **DataFrame** es la estructura con la que Pandas manipula datos. Es un equivalente logico a una tabla de Excel, o una tabla de SQL, con filas y columnas. 

## Lectura de datos de COVID desde un archivo csv

```python
df_datosCovid = pd.read_csv("datos/mexico_covid19.csv")
df_datosCovid.head()
```

## Exploracion de los datos con `describe()`

```python
df_datosCovid.describe()
```

## Exploracion de las columnas con `dtypes`

## Seleccionar un grupo de columnas del DataFrame original e imprimir estadistica descriptiva

```python
df_Filtrado = df_datosCovid[["ORIGEN", "FECHA_ACTUALIZACION", "PAIS_ORIGEN", "SEXO", "EMBARAZO", "EDAD", "DIABETES", "EPOC", "ASMA"]]
df_Filtrado.describe()
```

## Seleccionar un grupo de filas basado en una condicion logica

```python
df_Filtrado[df_Filtrado["SEXO"] == 2].describe()
```

## Seleccionar un grupo de filas basado en multiples condiciones logicas

```python
# Todos los hombres con edad entre 40 y 60
# (Condicion1) & (Condicion2) => Condicion1 and Condicion2
# (Condicion1) | (Condicion2) => Condicion1 or Condicion2
df_Filtrado[(df_Filtrado["SEXO"] == 2) & (df_Filtrado["EDAD"] >= 40) & (df_Filtrado["EDAD"] <= 60)]
```
